```python
def random_items(question=None, answer=None, *other):
    """question輸入題目
    answer輸入答案
    *other輸入其他選項
    回傳完整題目與正確選項
    """
    import random
    
    choose = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
    answers = [i for i in other]
    answers.append(answer)
    random.shuffle(answers)
    answers_choose = answers.index(answer)
    complete_question = str(question) + '\n\t'
    for i in range(len(answers)):
        complete_question += '(' + choose[i] + ')' +answers[i] + '\n\t'
    return complete_question, choose[answers_choose]
```

```python
test_questions = {}

k, v = random_items("「潯陽江頭夜送客」下一句？", "楓葉荻花秋瑟瑟", "主人下馬客在船", "唯見江心秋月白")
test_questions[k] = v

k, v = random_items("再下一句？", "主人下馬客在船", "大珠小珠落玉盤", "忽聞水上琵琶聲")
test_questions[k] = v

k, v = random_items("再下一句？", "舉酒欲飲無管絃", "嘈嘈切切錯雜彈", "忽聞水上琵琶聲")
test_questions[k] = v

k, v = random_items("再下一句？", "醉不成歡慘將別", "大珠小珠落玉盤", "忽聞水上琵琶聲")
test_questions[k] = v

k, v = random_items("再下一句？", "別時茫茫江浸月", "大珠小珠落玉盤", "忽聞水上琵琶聲")
test_questions[k] = v

k, v = random_items("再下一句？", "忽聞水上琵琶聲", "大珠小珠落玉盤", "忽聞水上琵琶聲")
test_questions[k] = v

k, v = random_items("再下一句？", "主人忘歸客不發", "嘈嘈切切錯雜彈", "忽聞水上琵琶聲")
test_questions[k] = v

k, v = random_items("再下一句？", "尋聲暗問彈者誰", "大珠小珠落玉盤", "忽聞水上琵琶聲")
test_questions[k] = v

k, v = random_items("再下一句？", "琵琶聲停欲語遲", "大珠小珠落玉盤", "忽聞水上琵琶聲")
test_questions[k] = v

k, v = random_items("再下一句？", "移船相近邀相見，添酒回燈重開宴", "輕攏慢撚抹復挑，初為霓裳後綠腰", "沉吟放撥插絃中，整頓衣裳起斂容")
test_questions[k] = v
```

```python
import time

print('請依序選出句子，接龍成一篇完整的〈琵琶行〉')

start=time.time()

score=0
for i in test_questions.keys():
    print('\n',i)
    answer = str.upper(input('答案：'))
    if answer==test_questions[i]:
        print('\t答對')
        score+=1
    else:
        print('\t答錯')
end=time.time()
total=end-start
print('總共耗時：%.5f秒,得分：%d'%(total,score))

notquit = True
while notquit:
    if str.lower(input("輸入q後按Enter結束程式")) == 'q': notquit = False
```
